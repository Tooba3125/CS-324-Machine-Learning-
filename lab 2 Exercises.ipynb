# Q 3

The range in the pd.cut function works by dividing the numeric column into intervals (or "bins") and assigning each value in the column to one of these intervals based on its magnitude.

pd.cut(column, bins, labels)

```python
df['Status'] = pd.cut(
    df['Cases'],
    bins=[0, 3_000_000, 4_000_000, float('inf')],
    labels=['Low', 'Medium', 'High']
)
df
```
